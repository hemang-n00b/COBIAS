```python
import pandas as pd
df = pd.DataFrame()
```

```python
with open('data.txt', 'r') as file:
    sentences = file.readlines()

sentences = [sentence.strip() for sentence in sentences]
sentences
```

```python
# Column 1: ID column which is test_0, test_1, ...
df['id'] = ['test_' + str(i) for i in range(len(sentences))]
df['sentence'] = sentences

# Target term column should be blank
df['target_term'] = ''
df['context_points'] = ''
df
```

```python
# Load original dataset to get context_points
data_original = pd.read_csv('/Users/priyanshul/Documents/COBIAS-Hemang/datasets/COBIAS.csv')
data_original = data_original[['sentence', 'context_points']]
data_original
```

```python
for i, row in df.iterrows():
    try:
        context_points = data_original[data_original['sentence'] == row['sentence']]['context_points'].values[0]
        df.at[i, 'context_points'] = context_points
    except:
        continue

df
```

```python
# Remove rows where context_points is empty
df = df[df['context_points'] != '']
df['id'] = ['test_' + str(i) for i in range(len(df))]
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('data.csv', index=False)
```
