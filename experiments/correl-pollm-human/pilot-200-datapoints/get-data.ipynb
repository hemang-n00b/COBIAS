```python
import pandas as pd
df = pd.read_csv("../../../datasets/DATASET.csv")
df.head()
```

```python
df = df[df["original_dataset"] == "cs"]
len(df)
```

```python
# Sample 200 datapoints
df = df.sample(200)
df.head()
```

```python
sentences = []
for i, row in df.iterrows():
    if row["direction"] == "stereo":
        sentences.append(row["sentence_1"])
    else:
        sentences.append(row["sentence_2"])

len(sentences)
```

```python
with open("data.txt", "w") as f:
    for sentence in sentences:
        f.write(sentence + "\n")
```

```python
# Make df with one column being these sentences and other columns being "annot_1", "annot_2" being empty
df = pd.DataFrame(sentences, columns=["sentence"])
df["annot_1"] = ""
df["annot_2"] = ""

df.to_csv("data_annotated.csv", index=False, mode="x")
```

```python
data = pd.read_csv("data_annotated.csv")
data.head()
```

```python
cobias = pd.read_csv("~/Precog/UMBC-CA/results/evaluations.csv")
cobias.head()
```

```python
dataset = pd.read_csv("../../../datasets/DATASET.csv")

cobias_scores = []

for i, row in data.iterrows():
    sentence = row["sentence"]

    # Find id of this sentence in dataset -- it will be in either sentence_1 or sentence_2
    id = dataset[(dataset["sentence_1"] == sentence) | (dataset["sentence_2"] == sentence)].index[0]
    id = f"cs_{id}"
   
    # Find the corresponding row in cobias
    cobias_row = cobias[cobias["id"] == id]
    
    cobias_scores.append(cobias_row["final_score"])
        
```

```python
temp = []
for i in range(200):
    print(i)
    try:
        temp.append(cobias_scores[i].tolist()[0])
    except:
        temp.append('')
# cobias_scores
```
