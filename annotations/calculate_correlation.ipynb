```python
from scipy.stats import spearmanr
import pandas as pd 
import numpy as np 

df = pd.read_csv("annotations.csv")
df2 = pd.read_csv("cobias_scores.csv")

score1 = df2["cobias"].tolist()
score2 = df[["Annotator 1" ,"Annotator 2" , "Annotator 3" ]].mean(axis=1).tolist()
score2
```

```python
correlation, p_value = spearmanr(score1, score2)
print("Spearman correlation coefficient:", correlation)
print("p-value:", p_value)
```

```python
from statsmodels.stats.inter_rater import fleiss_kappa, aggregate_raters

# Get scores for each annotator
annotators = df[["Annotator 1" ,"Annotator 2" , "Annotator 3" ]].values

# Aggregate scores
agg_scores, _ = aggregate_raters(annotators)

# Compute Fleiss' Kappa
kappa = fleiss_kappa(agg_scores, method='fleiss')
```

```python
# Calculate krippendorff's alpha
from krippendorff import alpha

annotators = df[["Annotator 1" ,"Annotator 2" , "Annotator 3" ]].values

# Transpose
annotators = annotators.T

alpha(annotators)
```
